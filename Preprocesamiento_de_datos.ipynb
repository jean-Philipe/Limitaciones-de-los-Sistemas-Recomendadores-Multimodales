Extracción de varios embeddings

````python
from transformers import AutoModelForCausalLM, AutoProcessor, GenerationConfig
from PIL import Image
import pandas as pd
import torch
import numpy as np
import requests
import os

# Cargar el procesador y el modelo
processor = AutoProcessor.from_pretrained(
    'allenai/MolmoE-1B-0924',
    trust_remote_code=True,
    torch_dtype='auto',
    device_map='auto'
)

model = AutoModelForCausalLM.from_pretrained(
    'allenai/MolmoE-1B-0924',
    trust_remote_code=True,
    torch_dtype='auto',
    device_map='auto'
    offload_folder="offload_weights"
)

# Función para extraer embeddings con imágenes y texto
def extract_molmo_embeddings(csv_path, image_column, text_column, window_i, window_f, save_interval=1000, start_iteration=0):
    """
    Extrae embeddings usando MolmoE-1B de un CSV con columnas de imágenes y textos.

    Argumentos:
    - csv_path: Ruta al archivo CSV con datos.
    - image_column: Nombre de la columna que contiene las URLs de imágenes.
    - text_column: Nombre de la columna que contiene el texto asociado.
    - window_i: Porcentaje inicial del rango de datos (0 a 1).
    - window_f: Porcentaje final del rango de datos (0 a 1).
    - save_interval: Frecuencia de guardado en filas procesadas.
    - start_iteration: Iteración desde la que se debe reanudar.
    """
    # Leer el archivo CSV
    data = pd.read_csv(csv_path)

    # Calcular el rango de filas basado en los porcentajes
    start_index = int(len(data) * window_i)
    end_index = int(len(data) * window_f)
    data = data.iloc[start_index:end_index]
    print(f"Procesando rango [{start_index}, {end_index}], total filas: {len(data)}")

    # Crear directorio para los embeddings
    os.makedirs("molmo_embeddings", exist_ok=True)

    embeddings = {}

    # Iterar por las filas del rango seleccionado
    for i, (_, row) in enumerate(data.iterrows(), start=start_iteration):
        # Saltar iteraciones procesadas
        if i < start_iteration:
            continue

        # Obtener imagen y texto
        image_url = row[image_column]
        text = row[text_column]

        try:
            # Procesar la imagen
            image = Image.open(requests.get(image_url, stream=True).raw)
            inputs = processor.process(images=[image], text=text)

            # Mover inputs al dispositivo correcto
            inputs = {k: v.to(model.device).unsqueeze(0) for k, v in inputs.items()}

            # Generar embeddings
            with torch.no_grad():
                output = model.generate_from_batch(
                    inputs,
                    GenerationConfig(max_new_tokens=200, stop_strings="<|endoftext|>"),
                    tokenizer=processor.tokenizer
                )

            # Extraer tokens generados y decodificar
            generated_tokens = output[0, inputs['input_ids'].size(1):]
            generated_text = processor.tokenizer.decode(generated_tokens, skip_special_tokens=True)

            # Guardar embeddings generados
            embeddings[row['item_id']] = generated_text

            # Guardado parcial
            if (i + 1) % save_interval == 0:
                save_name = f"molmo_embeddings/embeddings_part_{i // save_interval}.npy"
                np.save(save_name, embeddings)
                print(f"Guardado parcial en iteración {i + 1}: {save_name}")
                embeddings.clear()

        except Exception as e:
            print(f"Error procesando fila {i}: {e}")

    # Guardar los embeddings restantes
    if embeddings:
        save_name = f"molmo_embeddings/embeddings_final.npy"
        np.save(save_name, embeddings)
        print(f"Guardado final completado: {save_name}")

    return embeddings

# Ejemplo de uso
csv_path = "item_info.csv"  # Archivo CSV con columnas de URL de imágenes y texto
extract_molmo_embeddings(csv_path, image_column="image_url", text_column="description", window_i=0, window_f=1, save_interval=1000)
````

### Embeddings de texto

````python
import clip
import torch
import numpy as np
import pandas as pd

# Load the model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Max tokens CLIP supports
MAX_TOKENS = 77

# Function to truncate text based on token limit of 77
def truncate_text(text, max_tokens=MAX_TOKENS):
    # Tokenize the text
    tokenized_text = clip.tokenize([text], truncate=True).to(device)  # Ensure it's moved to the correct device
    num_tokens = tokenized_text.shape[1]  # Get the number of tokens (second dimension)

    # If the number of tokens exceeds the max allowed, truncate
    if num_tokens > max_tokens:
        # Truncate to the first max_tokens tokens (excluding special tokens)
        truncated_text = " ".join(text.split()[:max_tokens - 2])  # Keep space for special tokens
        tokenized_text = clip.tokenize([truncated_text]).to(device)  # Re-tokenize after truncating text

    return tokenized_text

# Function to get text embeddings
def extract_text_embeddings(csv_path, window_i, window_f, save_interval=1000):
    # Read the CSV file
    item_info = pd.read_csv(csv_path)

    # Create a dictionary to store the embeddings
    text_embeddings = {}

    # Get the range of rows based on the percentage
    start_index = int(len(item_info) * window_i)
    end_index = int(len(item_info) * window_f)
    item_info = item_info.iloc[start_index:end_index]
    print(f"Processing range [{start_index}, {end_index}], total rows: {len(item_info)}")

    # Process each row in the CSV
    for i, (_, row) in enumerate(item_info.iterrows()):
        row_embeddings = {}

        # Process specific columns: "tag", "title", "description"
        for column in ["tag", "title", "description"]:
            if column in row and pd.notna(row[column]):
                # Truncate the text and get the tokenized text
                tokenized_text = truncate_text(row[column])

                # Generate embeddings
                with torch.no_grad():
                    text_embedding = model.encode_text(tokenized_text)

                # Normalize the embedding and store it
                text_embedding /= text_embedding.norm(dim=-1, keepdim=True)
                row_embeddings[column] = text_embedding.cpu().numpy()

        # Store the embeddings for each row
        text_embeddings[row['item_id']] = row_embeddings

        # Periodically save embeddings
        if (i + 1) % save_interval == 0:
            save_name = f"text_embeddings[{start_index}:{end_index}]_part_{i // save_interval}.npy"
            np.save(save_name, text_embeddings)
            print(f"Partial save at iteration {i + 1}: {save_name}")
            text_embeddings.clear()

    # Save remaining embeddings after the loop
    if text_embeddings:
        save_name = f"text_embeddings[{start_index}:{end_index}]_final.npy"
        np.save(save_name, text_embeddings)
        print(f"Final save completed: {save_name}")

    return text_embeddings

# Example usage
csv_path = "item_info.csv"  # Path to your CSV file
text_embeddings = extract_text_embeddings(csv_path, 0, 1, 1000)
````

````python
import numpy as np
import os

def merge_text_embeddings(output_pattern, final_filename):
    """
    Une múltiples archivos de embeddings generados en partes y los guarda en un único archivo.

    Args:
        output_pattern (str): Patrón de los nombres de los archivos de las partes (por ejemplo, "text_embeddings").
        final_filename (str): Nombre del archivo final combinado.

    Returns:
        dict: Diccionario combinado con todos los embeddings.
    """
    # Obtener la lista de archivos que coinciden con el patrón
    part_files = [f for f in os.listdir() if f.startswith(output_pattern) and f.endswith(".npy")]

    if not part_files:
        print("No se encontraron archivos para combinar.")
        return

    file_order = {}
    for i in range(len(part_files)):
      if part_files[i].endswith("final.npy"):
        file_order[len(part_files)-1] = part_files[i]
      if "text_embeddings[0:82865]_part_" in part_files[i]:
        file_order[int(part_files[i].strip(".npy").split("text_embeddings[0:82865]_part_")[1])] = part_files[i]

    combined_embeddings = {}

    # Cargar y combinar todos los archivos
    for i in range(len(part_files)):
        part_file = file_order[i]
        print(f"Cargando {part_file}...")
        part_embeddings = np.load(part_file, allow_pickle=True).item()
        combined_embeddings.update(part_embeddings)  # Combinar los diccionarios

    # Guardar el archivo combinado
    np.save(final_filename, combined_embeddings)
    print(f"Embeddings combinados guardados en {final_filename}")

    return combined_embeddings

# Ejemplo de uso
final_embeddings = merge_text_embeddings("text_embeddings", "combined_text_embeddings.npy")
````

### Embeddings de imágenes

````python
import os
import torch
import clip
import pandas as pd
from PIL import Image
import numpy as np

# Cargar el modelo CLIP
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Función para obtener embeddings de imágenes en una carpeta con información en un archivo CSV
def extract_image_embeddings(image_folder, csv_path, window_i, window_f, save_interval=100):
    # Leer el archivo CSV
    item_info = pd.read_csv(csv_path)

    # Crear un diccionario para almacenar los embeddings
    embeddings = {}

    # Obtenemos una ventana con window correspondiente al percentil que se revisa
    start_index = int(len(item_info) * window_i)
    end_index = int(len(item_info) * window_f)
    if window_f == 1:
        item_info = item_info.iloc[start_index:]
        print(f"Se procesa el rango [{start_index}:], el largo es {len(item_info)}")
    else:
        item_info = item_info.iloc[start_index:end_index]

    print(f"Se procesa el rango [{start_index}, {end_index}]")

    # Recorrer cada fila del archivo CSV
    for i, (_, row) in enumerate(item_info.iterrows()):
        image_name = row['item_id'] + ".jpg"  # Nombre de la imagen en el CSV
        image_path = os.path.join(image_folder, image_name)

        # Comprobar si el archivo de imagen existe
        if os.path.exists(image_path):
            # Cargar y procesar la imagen
            image = Image.open(image_path).convert("RGB")
            image_input = preprocess(image).unsqueeze(0).to(device)

            # Extraer el embedding de la imagen
            with torch.no_grad():
                image_embedding = model.encode_image(image_input)

            # Normalizar el embedding y almacenarlo en el diccionario
            image_embedding /= image_embedding.norm(dim=-1, keepdim=True)
            embeddings[image_name] = image_embedding.cpu().numpy()  # Convertir a numpy para facilitar su manejo
        else:
            print(f"Warning: La imagen '{image_name}' no se encontró en '{image_folder}'")
            print(image_path)

        # Guardar cada 'save_interval' iteraciones
        if (i + 1) % save_interval == 0:
            np.save(f"image_embeddings[{start_index}:{end_index}]_part_{i // save_interval}.npy", embeddings)
            print(f"Guardado parcial en iteración {i + 1}")
            embeddings.clear()  # Vaciar el diccionario para el siguiente bloque de embeddings

    # Guardar cualquier embedding restante después del bucle
    if embeddings:
        np.save(f"image_embeddings[{start_index}:{end_index}]_final.npy", embeddings)
        print("Guardado final completado.")

    return embeddings

# Llamada a la función y ejemplo de uso
image_folder = "cover"  # Carpeta con imágenes
csv_path = "item_info.csv"  # Archivo CSV con información de cada imagen

image_embeddings = extract_image_embeddings(image_folder, csv_path, 0.66, 1, 1000)
````

````python
import torch
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from PIL import Image
import pandas as pd
import numpy as np

# Transformaciones para las imágenes
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Dataset personalizado para PixelRec
class PixelRecDataset(Dataset):
    def __init__(self, df, img_dir, transform=None):
        self.df = df
        self.img_dir = img_dir
        self.transform = transform
        self.user_item_dict = self._create_user_item_dict()
        self.all_items = list(self.df['item_id'].unique())

    def __len__(self):
        return len(self.df)

    def _create_user_item_dict(self):
      # Crear un diccionario que mapea usuarios a ítems con los que han interactuado
      user_item_dict = {}
      for _, row in self.df.iterrows():
          user, item = row['user_id'], row['item_id']
          if user not in user_item_dict:
              user_item_dict[user] = set()
          user_item_dict[user].add(item)
      return user_item_dict

    def _get_image(self, item_id):
      img_path = f"{self.img_dir}/{item_id}.jpg"
      image = Image.open(img_path).convert('RGB')
      if self.transform:
          image = self.transform(image)
      return image

    def __getitem__(self, idx):
        anchor_row = self.df.iloc[idx]
        anchor_user = anchor_row['user_id']
        anchor_item = anchor_row['item_id']

        # Obtener una imagen de "anchor"
        anchor_image = self._get_image(anchor_item)

        # Seleccionar un "positive" aleatorio (otro ítem con el que el usuario ha interactuado)
        positive_item = np.random.choice(list(self.user_item_dict[anchor_user] - {anchor_item}))
        positive_image = self._get_image(positive_item)

        # Seleccionar un "negative" aleatorio (un ítem con el que el usuario no ha interactuado)
        negative_item = np.random.choice(list(set(self.all_items) - self.user_item_dict[anchor_user]))
        negative_image = self._get_image(negative_item)

        return {
            'anchor': anchor_image,
            'positive': positive_image,
            'negative': negative_image
        }
````

````python
class VisualEmbeddingModel(nn.Module):
    def __init__(self):
        super(VisualEmbeddingModel, self).__init__()
        self.model = models.resnet50(pretrained=True)
        self.model.fc = nn.Identity()  # Eliminar la capa final para obtener el embedding

    def forward(self, x):
        return self.model(x)

# Crear el modelo y enviar a GPU si está disponible
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
embedding_model = VisualEmbeddingModel().to(device)
````

````python
class TripletNet(nn.Module):
    def __init__(self, embedding_model):
        super(TripletNet, self).__init__()
        self.embedding_model = embedding_model

    def forward(self, anchor, positive, negative):
        anchor_emb = self.embedding_model(anchor)
        positive_emb = self.embedding_model(positive)
        negative_emb = self.embedding_model(negative)
        return anchor_emb, positive_emb, negative_emb

# Instanciar el modelo de triplet y definir la función de pérdida
triplet_model = TripletNet(embedding_model).to(device)
criterion = nn.TripletMarginLoss(margin=1.0, p=2)
optimizer = torch.optim.Adam(triplet_model.parameters(), lr=0.0001)
````

````python
import time
for epoch in range(1):
    start_time = time.time()
    triplet_model.train()
    running_loss = 0.0
    for data in dataloader:
        anchor = data['anchor'].to(device)
        positive = data['positive'].to(device)
        negative = data['negative'].to(device)

        optimizer.zero_grad()
        anchor_emb, positive_emb, negative_emb = triplet_model(anchor, positive, negative)

        loss = criterion(anchor_emb, positive_emb, negative_emb)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        print("Running_loss {}, Time_per_data: {}".format(running_loss, time.time()-start_time))


    print(f"Epoch {epoch+1}, Loss: {running_loss/len(dataloader)}")
````
